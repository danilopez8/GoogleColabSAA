##actividades canarias

```python
import pandas as pd
import numpy as np
```

```python
canarias=pd.read_csv('/content/sample_data/democanarias.csv')
```

```python
canarias
```

```python
canarias = canarias.set_index('Island')
canarias.loc[['El Hierro', 'La Gomera']]
```

```python
canarias.loc[:, ['Island','Province']]
```

```python
canarias.iloc[[0, 2, 4, 6], [1]]
```

```python
canarias.loc[canarias['Area'] > 1000]
```

```python
canarias['Ayuda'] = np.where(
    canarias['Area'] < 1000,
    canarias['Population'] * 0.30,   # Si el rea < 1000  30%
    canarias['Population'] * 0.20    # Si el rea  1000  20%
)

canarias
```

```python
ayudas = canarias.pivot_table(
    values='Ayuda',
    index='Province',
    aggfunc='sum'
)

ayudas
```
