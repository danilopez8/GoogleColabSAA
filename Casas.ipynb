```python
from sklearn.preprocessing import OneHotEncoder
# Seleccionar columnas categóricas
listacat = casas.select_dtypes(include='object').columns
casas_one = casas[listacat].fillna("Desconocido").astype(str)

# Crear el codificador
codificador = OneHotEncoder()

# Ajustar y transformar los datos
codificacion = codificador.fit_transform(casas_one)

# Obtener nombres de las nuevas columnas
nuevascol = codificador.get_feature_names_out(listacat)

# Crear DataFrame con las nuevas columnas codificadas
nuevosdatos = pd.DataFrame(codificacion.toarray(), columns=nuevascol)

nuevosdatos.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================
# Detectar outliers y borrarlos. Utilizar rango intercuartílico.
# ============================================================

# Solo columnas numéricas (float o int)
num_cols = casas.select_dtypes(include=["int64", "float64"]).columns

# Mostrar boxplot inicial (antes de eliminar outliers)
plt.figure(figsize=(10,5))
sns.boxplot(data=casas[num_cols])
plt.title("Distribución de variables numéricas ANTES de eliminar outliers")
plt.xticks(rotation=45)
plt.show()

# Calcular Q1, Q3 y IQR solo para las columnas numéricas
Q1 = casas[num_cols].quantile(0.25)
Q3 = casas[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtrar filas fuera del rango permitido
casas_sin_outliers = casas[
    ~((casas[num_cols] < (Q1 - 1.5 * IQR)) | (casas[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)
]

# Mostrar resultados
print("Filas originales:", len(casas))
print("Filas sin outliers:", len(casas_sin_outliers))
print("Filas eliminadas por outliers:", len(casas) - len(casas_sin_outliers))

# Mostrar boxplot después de eliminar outliers
plt.figure(figsize=(10,5))
sns.boxplot(data=casas_sin_outliers[num_cols])
plt.title("Distribución de variables numéricas DESPUÉS de eliminar outliers")
plt.xticks(rotation=45)
plt.show()

# (Opcional) Comparar histogramas antes vs después para cada variable
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(casas[col], color="salmon", label="Antes", kde=True, stat="density")
    sns.histplot(casas_sin_outliers[col], color="skyblue", label="Después", kde=True, stat="density")
    plt.title(f"Distribución de {col} antes y después de eliminar outliers")
    plt.legend()
    plt.show()

# Guardar el dataset sin outliers
casas_sin_outliers.to_csv("casas_sin_outliers.csv", index=False)
print("\n Dataset sin outliers guardado como 'casas_sin_outliers.csv'")
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# ===============================================
# Crear el nuevo dataset preprocesado
# (Usamos el dataset limpio sin outliers)
# ===============================================
dataset_preprocesado = casas_sin_outliers.copy()

# Comprobamos que no queden valores nulos
print("Valores nulos totales:", dataset_preprocesado.isnull().sum().sum())

# ============================
# Estandarizar con StandardScaler
# ============================

# Seleccionamos las columnas numéricas para escalar
num_cols = dataset_preprocesado.select_dtypes(include=["int64", "float64"]).columns

# Creamos el objeto escalador
scaler = StandardScaler()

# Ajustamos y transformamos los datos numéricos
dataset_preprocesado[num_cols] = scaler.fit_transform(dataset_preprocesado[num_cols])

# Mostrar un resumen del dataset estandarizado
print("\n Dataset estandarizado correctamente.")
print(dataset_preprocesado.head())

# ============================
# Guardar el dataset preprocesado final
# ============================
dataset_preprocesado.to_csv("casas_preprocesadas.csv", index=False)
print("\n Dataset final guardado como 'casas_preprocesadas.csv'")
```
