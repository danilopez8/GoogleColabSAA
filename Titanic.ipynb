##EJERCICIO 15

```python
# Filtrar los menores de edad
menores = titanic[titanic["Age"] < 18]

# Calcular el porcentaje de supervivientes (Survived == 1) por clase
porcentaje_por_clase = menores.groupby("Pclass")["Survived"].mean() * 100

print("Porcentaje de menores de edad que sobrevivieron en cada clase:")
print(porcentaje_por_clase)

#CON BUCLE FOR
print("Porcentaje de menores de edad que sobrevivieron en cada clase:\n")

for clase in sorted(titanic["Pclass"].unique()):
    menores_clase = titanic[(titanic["Pclass"] == clase) & (titanic["Age"] < 18)]
    porcentaje = menores_clase["Survived"].mean() * 100
    print(f"Clase {clase}: {porcentaje:.2f}%")
```

##EJERCICIO 16

```python
# Función que clasifica según la edad
def clasificar_edad(edad):
    if edad < 12:
        return "Niño"
    elif edad < 18:
        return "Adolescente"
    elif edad < 60:
        return "Adulto"
    else:
        return "Anciano"

# Crear nueva columna 'AgeGroup' aplicando la función
titanic["AgeGroup"] = titanic["Age"].apply(clasificar_edad)

# Verificar
titanic[["Age", "AgeGroup"]].head(10)
```

##EJERCICIO 17

```python
ancianos = titanic[titanic["AgeGroup"] == "Anciano"]

total_ancianos = len(ancianos)
ancianos_sobrevivientes = ancianos[ancianos["Survived"] == 1].shape[0]

print("Total de ancianos:", total_ancianos)
print("Ancianos que sobrevivieron:", ancianos_sobrevivientes)
```
