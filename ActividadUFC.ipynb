```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#solicitamos la p√°gina de la URL que se va a scrapear.
url='https://www.ufc.com/rankings'
#Hacemos la solicitud
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
datos = []
```

```python
# Encontrar todas las divisiones (categor√≠as)
divisiones = soup.find_all("div", class_="view-grouping")
```

```python
for division in divisiones:
    header = division.find("div", class_="view-grouping-header")
    if not header:
        continue

    categoria = header.get_text(strip=True)
    peleadores = division.find_all("a", href=True)

    for p in peleadores:
        nombre = p.get_text(strip=True)
        if nombre:
            datos.append({"Categoria": categoria, "Nombre": nombre})
```

```python
df=pd.DataFrame(datos)
```
